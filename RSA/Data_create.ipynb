```python
import pandas as pd
import random
from sympy import isprime, nextprime

# Function to generate a 64-bit prime number
def generate_64bit_prime():
    while True:
        candidate = random.getrandbits(32)
        candidate |= (1 << 32) | 1  # Ensure it's 64-bit and odd
        if isprime(candidate):
            return candidate

# Generate 300 pairs of 64-bit primes and their product
data = []
for _ in range(10000):
    p = str(generate_64bit_prime())
    q = str(generate_64bit_prime())
    n = int(p) * int(q)
    # print(p , "    ", q, "     ", n)
    data.append(("a"+str(p), "a"+str(q), "a"+str(n)))
    
print(data)
# Save to CSV as text
df = pd.DataFrame(data, columns=["p", "q", "n"])
file_path = "rsa_primes_32.csv"
df.to_csv(file_path, index=False, quoting=1)  # quoting=1 -> csv.QUOTE_ALL

file_path
```
